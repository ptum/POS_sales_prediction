```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
#### load surranding data 
surranding_data =pd.read_json('data/Surroundings.json',orient='table')

#### build a dataframe for surranding features of each store
surrandings_df = []
for k,x in surranding_data.values:
    for i in x:
        surrandings_df.append([k,i, (x[i]!=[])*1])
surrandings_df = pd.DataFrame.from_records(surrandings_df,columns=['store_id','surranding','values'])
len(surrandings_df.surranding.unique()) 
surrandings_df[surrandings_df.duplicated()]
surrandings = surrandings_df.pivot_table(index='store_id',columns='surranding',values='values',aggfunc='mean')
```

```python
surrandings_df.head()
```

```python
#### load sales data
sales = pd.read_csv('data/sales_granular.csv',index_col=0)
#### filling missing value
sales[sales.isnull()]=0


#### convert data to time seris format
datetime_data =pd.to_datetime(sales.columns)
sales.columns =datetime_data

sales_ts_df = sales.unstack().reset_index()
sales_ts_df.columns=['datetime','store_id','sales']

sales_ts_df['dayofweek']=sales_ts_df.datetime.dt.dayofweek
sales_ts_df['day']=sales_ts_df.datetime.dt.day
sales_ts_df['month']=sales_ts_df.datetime.dt.month
sales_ts_df['year']=sales_ts_df.datetime.dt.year

```

```python
#### calculate monthly sale as traget variable
monthly_sales = sales_ts_df.groupby(['store_id','year','month'])['sales'].sum().reset_index()
monthly_sales['date'] =map(lambda x:datetime.date(x[0],x[1],1), monthly_sales[['year','month']].values)

```

```python
monthly_sales.head()
```

```python
#### find the time interval with at least 50% of store having  sales>0
start_date = (monthly_sales[monthly_sales.sales>0].groupby(['date'])['sales'].count()/len(monthly_sales.store_id.unique())).reset_index()
start_date.plot.bar(x='date',y='sales')
plt.show()
selected_sales_ts=monthly_sales[monthly_sales.date.isin(start_date[start_date['sales']>0.5].date)]

print "Number of store in sales TS:",len(selected_sales_ts.store_id.unique())
```

```python
#### add surranding features to sales
sales_monthly_surranding = pd.merge(selected_sales_ts,surrandings_df,on='store_id',how='left')
#### drop stores without surranding features
sales_monthly_surranding.dropna(inplace=True)
print "Number of store in sales TS with Surranding features:",len(sales_monthly_surranding.store_id.unique())
```

```python
selected_sales_ts.groupby(['store_id'])['sales'].mean().plot.box()
plt.show()

selected_sales_ts.groupby(['store_id'])['sales'].mean().describe()
```

```python
####  Visual compare of sales for having a feature in surranding or not
sns.factorplot(data = sales_monthly_surranding, x = 'month', y = "sales", 
               row = 'surranding',col='values',) 
plt.show()
```

```python
#### Define target variables for stores having surranding features

store_index = selected_sales_ts.store_id[selected_sales_ts.store_id.isin(sales_monthly_surranding.store_id)].unique()

# average monthly sale value
target_variable = selected_sales_ts.groupby(['store_id']).sales.mean()[store_index].values
plt.boxplot(target_variable)
plt.show()

# binary class of being top seller store (average monthly sale>threshold) or not
threshold = 3016.875000
binary_target_variable  = (target_variable>threshold)*1

ax = sns.countplot(binary_target_variable,label="Count")      
plt.show()

```

```python
#### PCA of surranding variables 

from sklearn.decomposition import PCA

n=10
pca = PCA(n_components=n)

y = binary_target_variable
X = surrandings.loc[store_index].values

principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc'+str(i) for i in range(n)])
finalDf = principalDf


fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
ax.scatter(finalDf['pc1']
               , finalDf['pc2']
               , c = y
               , s = 50)

ax.grid()
plt.show()

plt.bar(range(n),pca.explained_variance_ratio_)
plt.show()
print "Explained variance: ",pca.explained_variance_ratio_

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_curve, auc

y = binary_target_variable
X = surrandings.loc[store_index].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
rf.fit(X_train, y_train)


y_pred = rf.predict(X_test)
y_score = rf.predict_proba(X_test)

fpr, tpr, thresholds = roc_curve(y_test,y_score[:,1],pos_label=1)
roc_auc = auc(fpr, tpr)

feature_score = rf.feature_importances_

all_names = suranding_names.tolist() +['average_montly_sales',]
colors = []
names = []
for i in xrange(len(feature_score)):
    if feature_score[i]>0.02:
        colors.append('r')
        names.append(all_names[i])
        selected_features['rf_clf'].append([all_names[i],feature_score[0]])
    else:
        colors.append('b')
        names.append('')
    
plt.bar(range(len(feature_score)), feature_score,color=colors)
plt.xticks(range(len(names)),names,rotation=70)
plt.title('AUC- ROC:'+str(roc_auc))

plt.show()
plt.boxplot(feature_score)
plt.show()
```
